```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from collections import Counter

# 网页链接
url = "https://zh.wikipedia.org/zh-tw/%E6%88%B4%E8%B3%87%E7%A9%8E"

# 发送请求获取页面内容
response = requests.get(url)
html_content = response.content

# 使用BeautifulSoup解析页面内容
soup = BeautifulSoup(html_content, 'html.parser')

# 找到目标表格，这里使用表格的类名 "wikitable" 作为标识
target_table_class = "wikitable"
target_table = soup.find('table', class_=target_table_class)

# 检查是否找到了目标表格
if target_table:
    # 读取表格数据
    df = pd.read_html(str(target_table))[0]

    # 删除指定列
    columns_to_drop = ["年份", "公開賽級別", "項目", "成績"]
    df = df.drop(columns=columns_to_drop, axis=1)

    # 将DataFrame转为文本形式
    table_text = df.to_string(index=False)

    # 打印处理后的文本内容
    print(table_text)
else:
    print("未找到目标表格")
```

```python
import re

# 假设你已经得到了处理后的文本 table_text

# 要删除的字词列表
words_to_remove = ["賽","亞洲","東亞","澳洲","比","奧林匹克","巡迴","總決","事","羽","球","大獎","運動會","公開","超級","黃金","大師","YONEX","TOYOTA","錦標","福州","系列","大學","世界","頂級","聯",]

for word in words_to_remove:
    table_text = table_text.replace(word, '')
    
# 将文本中的"臺"替换成"台"
table_text = table_text.replace("臺", "台")

# 将文本中的"全英"替换成"英國"
table_text = table_text.replace("全英", "英國")

# 将文本中的"台北"替换成"中華台北"
table_text = table_text.replace("中華台北", "台北")

# 打印处理后的文本
print(table_text)
```

```python
# 使用正则表达式提取国家名称
countries = re.findall(r'\b\w+台?\b', table_text)

# 使用Counter统计国家出现次数
country_counts = Counter(countries)

# 将结果分别存储在两个列表中
countries_list, counts_list = zip(*country_counts.items())

# 打印国家和对应的出现次数
print("Countries:", countries_list)
print("Counts:", counts_list)
```

```python
# 創建用於HeatMap的數據點
heat_data = []

for location, info in locations.items():
    # 使用 place 作為鍵來檢索重複次數
    count = word_freq.get(location, 0)
if count > 0:
    heat_data.append([info[0], info[1], count])

# 創建熱點圖層
HeatMap(heat_data).add_to(m)


# 顯示地圖
m
```
