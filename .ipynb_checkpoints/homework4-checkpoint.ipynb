```python
from pytube import YouTube
from googleapiclient.discovery import build

# 設定API金鑰，請替換為你自己的金鑰
API_KEY = "AIzaSyD5lCNFOrHcgIwU-aZG2wxA3_D7NQYz8mk"

# 影片的URL
video_url = "https://www.youtube.com/watch?v=ejWlcJ2HIzM&list=PLu48w5JmTaueymgOlwYzLph8p3ptCbdeL&index=25"

# 取得影片ID
video_id = YouTube(video_url).video_id

# 用YouTube API取得留言
youtube = build('youtube', 'v3', developerKey=API_KEY)

# 取得影片的留言串列
comments = []
results = youtube.commentThreads().list(part='snippet', videoId=video_id, textFormat='plainText').execute()

while results:
    for item in results['items']:
        comment = item['snippet']['topLevelComment']['snippet']['textDisplay']
        comments.append(comment)

    # 檢查是否還有下一頁的留言
    if 'nextPageToken' in results:
        results = youtube.commentThreads().list(part='snippet', videoId=video_id, textFormat='plainText',
                                                pageToken=results['nextPageToken']).execute()
    else:
        break

# 列印留言
for i, comment in enumerate(comments, start=1):
    print(f"{i}. {comment}")
```

```python
# 假設你已經有了留言的串列 comments

# 要搜尋的特定名詞
target_term = "紅隊"

# 儲存包含特定名詞的留言的串列
comments_with_term = []

# 遍歷留言串列，搜尋特定名詞
for comment in comments:
    if target_term in comment:
        comments_with_term.append(comment)

# 列印包含特定名詞的留言
for i, comment in enumerate(comments_with_term, start=1):
    print(f"{i}. {comment}")
```

```python
import re

# 假設你已經有了留言的串列 comments_with_term

# 定義一個函式，用於刪除標點符號和 emoji，以及其他特定字詞
def process_comment(comment):
    # 刪除標點符號
    comment = re.sub(r'[^\w\s]', '', comment)
    # 刪除 emoji
    comment = re.sub(r'[\U00010000-\U0010ffff]', '', comment)
    
    return comment

# 創建一個新的串列，存儲處理後的留言
processed_comments = []

# 遍歷留言串列，處理每條留言
for comment in comments_with_term:
    processed_comment = process_comment(comment)
    processed_comments.append(processed_comment)

# 列印處理後的留言
for i, processed_comment in enumerate(processed_comments, start=1):
    print(f"{i}. {processed_comment}")
```

```python
import nltk
nltk.download('punkt')  # 下載分詞所需的資源

from nltk.tokenize import word_tokenize

# 假設你已經有了處理後的留言的串列 processed_comments

# 創建一個新的串列，存儲拆分後的詞語
tokenized_comments = []

# 遍歷處理後的留言，進行詞語拆分
for processed_comment in processed_comments:
    tokens = word_tokenize(processed_comment)
    tokenized_comments.append(tokens)

# 列印拆分後的詞語
for i, tokens in enumerate(tokenized_comments, start=1):
    print(f"{i}. {tokens}")
```

```python
from collections import Counter

# 假設你已經有了詞語拆分後的串列 tokenized_comments

# 將所有詞語合併成一個列表
all_tokens = [token for tokens in tokenized_comments for token in tokens]

# 使用 Counter 進行統計
word_counts = Counter(all_tokens)

# 列印各字詞出現次數
for word, count in word_counts.items():
    print(f"{word}: {count}")
```

```python
from wordcloud import WordCloud

# 設定中文字體路徑
font_path = "C:/Windows/Fonts/mingliu.ttc"

# 初始化 WordCloud
wc = WordCloud(font_path=font_path, background_color="white", max_words=1000, width=1000, height=800)

# 生成文字雲
wordcloud = wc.generate_from_frequencies(word_counts)

# 顯示文字雲圖片
plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # 不顯示座標軸
plt.show()
```
